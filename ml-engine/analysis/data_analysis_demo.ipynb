```python
# Load and preprocess data for spam detection
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import re
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import nltk
nltk.download('stopwords')

# Use a sample dataset since emails copy.csv is missing
sample_data = {
    'Message': [
        'Congratulations! You have won a free ticket.',
        'Hi, can we reschedule our meeting?',
        'Urgent: Your account has been compromised.',
        'Lunch at 1pm?',
        'You are selected for a lottery prize.'
    ],
    'Category': ['spam', 'ham', 'spam', 'ham', 'spam']
}
df = pd.DataFrame(sample_data)

text_col = 'Message'

# Ensure category values are numeric 0/1
df['Category'] = df['Category'].str.strip().str.lower()
df.loc[df['Category'] == 'spam', 'Category'] = 1
df.loc[df['Category'] == 'ham', 'Category'] = 0
df['Category'] = pd.to_numeric(df['Category'], errors='coerce')

if not set(df['Category'].dropna().unique()).issubset({0, 1}):
    raise ValueError("Category values must be spam/ham or 0/1 after normalization")

# Normalize missing values
data = df.where(pd.notnull(df), '')

# Use this text column for features
X = data[text_col]
Y = data['Category'].astype('int')
```

```python
# Advanced text preprocessing, Naive Bayes model, and hyperparameter tuning
import re
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
nltk_stopwords = set(stopwords.words('english'))
stemmer = PorterStemmer()
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9 ]', '', text)
    words = text.split()
    words = [stemmer.stem(word) for word in words if word not in nltk_stopwords]
    return ' '.join(words)
# Clean all messages
data['Message_clean'] = data['Message'].apply(clean_text)
# Split data
X = data['Message_clean']
Y = data['Category']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
# Vectorize
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
# Naive Bayes model with hyperparameter tuning
nb_model = MultinomialNB()
param_grid = {'alpha': [0.1, 0.5, 1.0, 2.0]}
grid = GridSearchCV(nb_model, param_grid, cv=5)
grid.fit(X_train_features, Y_train)
best_nb_model = grid.best_estimator_
# Evaluate
train_pred = best_nb_model.predict(X_train_features)
test_pred = best_nb_model.predict(X_test_features)
train_acc = accuracy_score(Y_train, train_pred)
test_acc = accuracy_score(Y_test, test_pred)
print('Best Naive Bayes alpha:', grid.best_params_)
print('Train accuracy:', train_acc)
print('Test accuracy:', test_acc)
# Test on emails copy.csv and save output
test_df = pd.read_csv('emails copy.csv')
# Ensure labels are numeric (1=spam, 0=ham)
if 'Category' in test_df.columns:
    test_df['Category'] = test_df['Category'].map({'spam': 1, 'ham': 0}).fillna(test_df['Category'])

test_df['Message_clean'] = test_df['Message'].apply(clean_text)
test_features = feature_extraction.transform(test_df['Message_clean'])
test_predictions = best_nb_model.predict(test_features)
test_df['prediction'] = test_predictions
test_df['prediction_label'] = test_df['prediction'].map({1: 'spam mail', 0: 'ham mail'})
test_df.to_csv('emails_predictions_nb.csv', index=False)
print('Predictions saved to emails_predictions_nb.csv')
# Check if predictions match original labels
original_labels = test_df['Category']
predicted_labels = test_df['prediction']
comparison = original_labels == predicted_labels
print('Number of matches:', comparison.sum())
print('Total emails:', len(comparison))
print('Accuracy:', comparison.sum() / len(comparison))
```
